```python
df1 = df[['Created', 'Title', 'Title Type', 'IMDb Rating', 'Runtime (mins)', 'Year', 'Genres', 'Num Votes', 'Release Date', 'Directors']]
```

```python
df1
```

```python
df1 = df1.rename(columns={'Created': 'Adicionado', 'Title': 'Nome', 'Title Type': 'Tipo', 'IMDb Rating': 'Nota no IMDb', 'Runtime (mins)': 'Duração', 'Year': 'Ano de Lançamento', 'Genres': 'Gênero', 'Num Votes': 'Número de Votos', 'Release Date': 'Data de Lançamento', 'Directors': 'Diretores'})
```

```python
df1.head(5)
```

```python
# Títulos com menor duração (em minutos)
df1[['Nome', 'Duração']].sort_values(by='Duração').head()
```

```python
# Títulos com maior duração (em minutos)
df1[['Nome', 'Duração']].sort_values(by='Duração', ascending=False).head()
```

```python
# Menor nota no IMDb
df1[['Nome', 'Nota no IMDb']].sort_values(by='Nota no IMDb').head(5)
```

```python
# Maior nota no IMDb
df1[['Nome', 'Nota no IMDb']].sort_values(by='Nota no IMDb', ascending=False).head(5)
```

```python
# Capturando somente o gênero principal de cada filme.
# Adicionando à coluna GENRES os valores editados.
result = []
for i in df.Genres:
    split_string = i.split(",", 1)
    #print(split_string[0])
    result.append(split_string[0])
    

result

df["Genres"] = result
```
