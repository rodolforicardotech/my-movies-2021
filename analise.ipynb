```python
# Capturando somente o gênero principal de cada filme.
# Adicionando à coluna GENRES os valores editados.
result = []
for i in df.Genres:
    split_string = i.split(",", 1)
    #print(split_string[0])
    result.append(split_string[0])
    

result

df["Genres"] = result
```

```python
# Adicionando o tipo de data à coluna mês
df['Month'] = pd.DatetimeIndex(df1['Adicionado']).month
```

```python
# df['Month'].loc[104] = 12
```

```python
# criando um dataframe nova para a visualização dos filmes por mês
meses = df[["Month"]]
```

```python
# Criando um DataFrame a partir do VALUE COUNTS. IMPORTANTE!!!!!!!!
meses = meses["Month"].value_counts().rename_axis("Meses").reset_index(name="Contagem")
```

```python
# adicionando a organização pelos meses
meses.sort_values(by="Meses", inplace=True)
```

```python
# plotando o primeiro gráfico de linhas!
meses.plot(kind="line", x="Meses", xticks=range(1,13))
fig =  px.line(meses, meses["Meses"], meses["Contagem"], title="Filmes por Mês")
#fig.write_html('filmes_por_mes.html', auto_open=True)
```

```python
# Cinco melhores filmes
notas_maiores = df1[["Nome", "Nota no IMDb"]].sort_values(by="Nota no IMDb", ascending=False).head()
#fig = px.bar(x=notas_maiores["Nome"], y=notas_maiores["Nota no IMDb"])
fig =  px.bar(notas, notas_maiores["Nome"], notas_maiores["Nota no IMDb"], title="Top 5 - Filmes com as Maiores Notas")
#fig.write_html('maiores_notas.html', auto_open=True)
```

```python
notas_menores = df1[["Nome", "Nota no IMDb"]].sort_values(by="Nota no IMDb").head()
#fig = px.bar(x=notas_menores["Nome"], y=notas_menores["Nota no IMDb"])
fig =  px.bar(notas, notas_menores["Nome"], notas_menores["Nota no IMDb"], title="Top 5 - Filmes com as Menores Notas")
#fig.write_html('menores_notas.html', auto_open=True)
```
